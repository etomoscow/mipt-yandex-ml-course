```python
dat = df1[['latitude','longitude']]
dat
```

```python
clustering = MeanShift(bandwidth = 0.1, n_jobs = -1 ).fit(dat)
```

```python
ans = clustering.predict(dat)
ans
```

```python
from collections import Counter
Counter(ans)
```

```python
numb_clust = {}
for elements in ans:
    if elements in numb_clust:
        numb_clust[elements] += 1
    else:
        numb_clust[elements] = 0 
```

```python
true_clust = []
for elements in numb_clust:
    if numb_clust[elements] > 15:
        true_clust.append(elements)
```

```python
true_clust
```

```python
true_centers = clustering.cluster_centers_[true_clust]
```

```python
print(true_centers)
```

```python
office_adresses = [
    [33.751277, -118.188740],
    [25.867736, -80.324116],
    [51.503016, -0.075479],
    [52.378894, 4.885084],
    [39.366487, 117.036146],
    [-33.868457, 151.205134],
]
```

```python
def dist(x,y):
    temporary_array = [((x-coord[0]) ** 2 + (y-coord[1]) ** 2 )**(1/2) for coord in office_adresses]
    return(min(temporary_array))
```

```python
result = []
for elements in true_centers:
    a = dist(elements[0],elements[1])
    result.append([a, elements[0], elements[1]])
print(sorted(result)[0])
```
